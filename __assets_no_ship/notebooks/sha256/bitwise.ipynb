# **Bitwise Operations**

This is an overview of common bitwise operations, which are essential for understanding the functions used in the SHA family of hashing algorithms. 

```python
red = lambda s: f"\033[31m{s}\033[0m"
green = lambda s: f"\033[32m{s}\033[0m"
```

## Logic Operations

### `AND` Operation
*Are both bits 1?*
* Mathematical symbol $\land$
* Python symbol `&`

```python
x = 0b1
y = 0b0

print("Bitwise AND of two 1-bit binary numbers:\n")
print(f"{x} & {x} = {x & x:01b}")
print(f"{x} & {y} = {x & y:01b}")
print(f"{y} & {x} = {y & x:01b}")
print(f"{y} & {y} = {y & y:01b}")

print("\nBitwise AND of two 8-bit binary numbers:\n")

x = 0b10101010
y = 0b11001100
z = x & y

print(f"   {x:08b}")
print(f" & {y:08b}")
print("⎯⎯⎯⎯⎯⎯⎯")
print(f"   {z:08b}")
```

### `OR` Operation
*Is one of the bits 1?*
* Mathematical symbol $\lor$
* Python symbol `|`

```python
x = 0b1
y = 0b0

print("Bitwise OR of two 1-bit binary numbers:\n")
print(f"{x} | {x} = {x | x:01b}")
print(f"{x} | {y} = {x | y:01b}")
print(f"{y} | {x} = {y | x:01b}")
print(f"{y} | {y} = {y | y:01b}")

print("\nBitwise OR of two 8-bit binary numbers:\n")

x = 0b10101010
y = 0b11001100
z = x | y

print(f"   {x:08b}")
print(f" | {y:08b}")
print("⎯⎯⎯⎯⎯⎯⎯")
print(f"   {z:08b}")
```


### `XOR` Operation
*Are the bits different?*
* Mathematical symbol $\oplus$
* Python symbol `^`

```python
x = 0b1
y = 0b0

print(f"{x} ^ {x} = {x ^ x:01b}")
print(f"{x} ^ {y} = {x ^ y:01b}")
print(f"{y} ^ {x} = {y ^ x:01b}")
print(f"{y} ^ {y} = {y ^ y:01b}")

print("\nBitwise AND of two 8-bit binary numbers:")

x = 0b10101010
y = 0b11001100
z = x ^ y

print(f"   {x:08b}")
print(f" ^ {y:08b}")
print("⎯⎯⎯⎯⎯⎯⎯")
print(f"   {z:08b}")
```


### `NOT` Operation
*flips the bit; 1 $\rightarrow$ 0, 0 $\rightarrow$ 1*
* Mathematical symbol $\neg$
* Python symbol `~`


> **NOTE**: 
>
>*Python is a high-level language, which means it does not directly work with bits. It actually stores all values as integers (decimal numbering system). This means that the bitwise operations are performed on the integer values themselves, and the results are interpreted back to bit notation (binary).* 
> 
> *As a result the NOT operation is actually defined as `~x == -(x + 1)`. So `~1 == -2`, which in binary is an infinite number of 1's followed by a 0: 1...10. So in order to get correct binary interpretation, we need to only look at the relevant part of the bit. In this case we are working with numbers of size 1 (1-bit numbers), which means we only look at the last binary digit*
>
>*The way we can isolate this in Python is by using the `AND` operator, and a "mask" (a binary number consisting of only 1's).*
>
>`~1 = 11...110`  
>`~0b1 & 0b1 = 11...110 & 00...001 = 00...000 = 0b0`  
>`~0b0 & 0b1 = 00...000 & 00...001 = 00...001 = 0b1`  

```python
x = 0b1
y = 0b0

print(f"~{x} = {~x& 0b1:01b}")
print(f"~{y} = {~y& 0b1:01b}")


print("\nBitwise NOT of an 8-bit binary number:")

x = 0b10101010
print(f"~{x:08b} = {~x & 0b11111111:08b}")  # Mask to keep only the last 8 bits
```

## Shifts & Rotations


### Shift Left Operation
*Slides all bits to the left by &n& positions. New bits on the right are filled with 0. Bits that fall off the left end are discarded (lost).*
* Mathematical symbol $<<$
* Python symbol `<<`


```python
x = 0b10101010
original = f"   {x:08b}"
intermediate_step = f"{red('101')} 01010 {green('000')}"
result = f"    {(x << 3) & 0b11111111:08b}"

print(f"{'Original:':<{65}}{original}")
print(f"{'Remove 3 from the left, add 3 zeros to the right:':<{65}}{intermediate_step}")
print(f"{'Left shift by 3:':<{65}}{result}")
```


### Shift Right Operation
*Slides all bits to the right by &n& positions. New bits on the left are filled with 0. Bits that fall off the right end are discarded (lost).*
* Mathematical symbol $>>$
* Python symbol `>>`


```python
x = 0b10101010
n = 3
original = f"    {x:08b}"
intermediate_step = f"{green('000')} 10101 {red('010')}"
result = f" {(x >> n) & 0b11111111:08b}"

print(f"{'Original:':<{65}}{original}")
print(f"{f'Remove {n} from the left, add {n} zeros to the right:':<{65}}{intermediate_step}")
print(f"{f'Right shift by {n}:':<{65}}{result}")
```

### Rotate Right Operation
Also slides bits right by n positions, but instead of discarding the bits that fall off, they wrap around to the left side.

This operation is not explicitly implemented in Python, but can be achieved by using the shift operations. First a shift right operation is performed, this removes $n$ digits from the right side. These need to be added at the front, so a seperate operation takes the original input binary and shifts it in such a way that the $n$ digits on the right have been shifted over by the correct amount. 

This depends on the size of the binary (variable `s`). In this case we are using bytes (numbers with 8 bits), so `s=8`. 

> NOTE: *Similar to the `NOT` operator, a mask needs to be used to get back a bit in the correct size. We know beforehand that the final number will be `s` bits, so a mask of `((1 << s) - 1)` can be used.* 
>
> *`1 << s` will create a bit 10...0 with `s` number of 0's. subtracting 1 from that will return 11...1 with `s-1` number of 1's*

```python
x = 0b10010011
n = 2
s = 8 #size of the binary number in bits

original = f"            {x:0{s}b}"
intermediate_step_1 = f"          {green('00')}100100   {red('11')}"
intermediate_step_2 = f" {red('100100')}   11{green('000000')}"
result = f"          {(x >> n | x << (s - n))&((1 << s) - 1):0{s}b}"

print(f"{'Original:':<{55}}{original}")
print(f"{f'Remove {n} from the right, add {n} zeros to the left:':<{55}}{intermediate_step_1}")
print(f"{f'Remove {s-n} from the left, add {s-n} zeros to the right:':<{55}}{intermediate_step_2}")
print(f"{f'Combine results:':<{55}}{result}")
```

### Rotate Left Operation
Opposite of the Rotate Right Operation. Slides bits left by `n` positions, but instead of discarding the bits that fall off, they wrap around to the right side.


```python
x = 0b10010011
n = 2
s = 8 #size of the binary number in bits

original = f"            {x:0{s}b}"
intermediate_step_1 = f"           {red('10')} 010011{green('00')}"
intermediate_step_2 = f"              {green('000000')}10 {red('010011')}"
result = f"              {(x << n | x >> (s - n))&((1 << s) - 1):0{s}b}"

print(f"{'Original:':<{55}}{original}")
print(f"{f'Remove {n} from the left, add {n} zeros to the right:':<{55}}{intermediate_step_1}")
print(f"{f'Remove {s-n} from the right, add {s-n} zeros to the left:':<{55}}{intermediate_step_2}")
print(f"{f'Combine results:':<{55}}{result}")
```
